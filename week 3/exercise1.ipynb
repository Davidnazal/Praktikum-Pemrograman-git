```python
def copy_lines(source_file, target_file, search_terms):
    with open(source_file, 'r') as infile, open(target_file, 'w') as outfile:
        for line in infile:
            for term in search_terms:
                if term in line:
                    outfile.write(line)

copy_lines("Biodata.txt", "DataDiri.txt", ["Nama", "NPM", "Program studi", "Alamat"])
```
