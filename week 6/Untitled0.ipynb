```python
#Weekly Exercise 3
baris = int(input("Masukkan angka: "))
dasar = 1
total = 0

for y in range(baris):
    if y == 0:
        dasar = 1
    else:
        dasar = dasar * (baris - y) // y
    print(dasar, end=" ")
    total += dasar

print(f"\nJumlah angka dalam baris ke {baris} adalah: {total}")

```
