```python

num = (input("Masukkan NPM anda: "))
angka_genap = 0
angka_ganjil = 0

for angka in (num) :
  if int(angka) % 2 == 0:
    angka_genap += 1
  else:
    angka_ganjil += 1

print("Jumlah angka genap:", angka_genap)
print("Jumlah angka ganjil:", angka_ganjil)
```
